Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
# Pilih kolom-kolom bertipe kategorikal
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Membuat pie chart untuk setiap kolom kategorikal
for col in categorical_columns:
    # Menghitung persentase setiap kategori
    category_counts = df[col].value_counts(normalize=True) * 100

    # Membuat pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Percentage of {col}')
    plt.axis('equal')  # Agar pie chart berbentuk bulat
    plt.show()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
# Menentukan batas nilai untuk lulus
pass_threshold = 75

# Fungsi untuk menentukan pass atau fail
def pass_or_fail(score):
    return 'Pass' if score >= pass_threshold else 'Fail'

# Menambahkan kolom baru untuk status pass/fail untuk setiap pelajaran
df['math_status'] = df['math score'].apply(pass_or_fail)
df['reading_status'] = df['reading score'].apply(pass_or_fail)
df['writing_status'] = df['writing score'].apply(pass_or_fail)

# Menghitung jumlah siswa yang pass dan fail untuk setiap pelajaran
math_count = df['math_status'].value_counts()
reading_count = df['reading_status'].value_counts()
writing_count = df['writing_status'].value_counts()

# Menampilkan hasil
print("Math Score - Pass/Fail Counts:")
print(math_count)
print("\nReading Score - Pass/Fail Counts:")
print(reading_count)
print("\nWriting Score - Pass/Fail Counts:")
print(writing_count)
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
# Menambahkan kolom 'average score' sebagai rata-rata dari math, reading, dan writing
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Melihat sekilas data dengan rata-rata yang ditambahkan
print(df[['math score', 'reading score', 'writing score', 'average score']].head())

# Visualisasi rata-rata skor berdasarkan kategori tertentu menggunakan box plot
def plot_avg_score_by_category(column_name):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=column_name, y='average score', data=df, palette='muted')
    plt.title(f'Average Score by {column_name}', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45)  # Rotasi label jika diperlukan
    plt.show()

# Memvisualisasikan rata-rata skor terhadap beberapa kategori
plot_avg_score_by_category('gender')
plot_avg_score_by_category('race/ethnicity')
plot_avg_score_by_category('parental level of education')
plot_avg_score_by_category('lunch')
plot_avg_score_by_category('test preparation course')

# Menghitung korelasi nilai rata-rata dengan kategori (menggunakan encoding)
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Melakukan one-hot encoding pada variabel kategorikal
df_encoded = pd.get_dummies(df[categorical_columns], drop_first=True)

# Menambahkan kolom 'average score' ke dataframe yang telah diencode
df_encoded['average score'] = df['average score']

# Menghitung korelasi
correlation_matrix = df_encoded.corr()

# Menampilkan korelasi 'average score' dengan variabel kategorikal
print("Correlation with Average Score:")
print(correlation_matrix['average score'].sort_values(ascending=False))

# Visualisasi heatmap korelasi
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix[['average score']], annot=True, cmap='YlGnBu', cbar=True)
plt.title('Correlation of Average Score with Categorical Variables', fontsize=14, fontweight='bold')
plt.show()
````

Prompt: beri kode untuk melihat  distribusi nilai  per mata pelajaran secara mendetail menggunakan Kernel Density Estimation

````python
# Mengatur ukuran plot
plt.figure(figsize=(18, 6))

# Membuat KDE plot untuk setiap mata pelajaran
subjects = ['math score', 'reading score', 'writing score']
colors = ['#FFB5E8', '#B5EAD7', '#FFDAC1']  # Warna pastel

for i, subject in enumerate(subjects):
    plt.subplot(1, 3, i + 1)  # Membuat subplot 1x3
    sns.kdeplot(df[subject], shade=True, color=colors[i], alpha=0.8)
    plt.title(f'Distribution of {subject.capitalize()}', fontsize=14)
    plt.xlabel(subject.capitalize(), fontsize=12)
    plt.ylabel('Density', fontsize=12)

# Tampilkan plot
plt.tight_layout()
plt.show()
````

Prompt : beri kode untuk melihat hubungan antar mata pelajaran, apakah siswa dengan nilai tinggi di satu mata pelajaran juga  mendapatkan nilai tinggi di mata pelajaran lainnya

````python
# Membuat scatter plot antar mata pelajaran
sns.pairplot(df[['math score', 'reading score', 'writing score']], kind='reg', diag_kind='kde',
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.6}})
plt.suptitle('Relationship between Math, Reading, and Writing Scores', y=1.02, fontsize=16)
plt.show()

# Hitung korelasi antar mata pelajaran
corr_matrix = df[['math score', 'reading score', 'writing score']].corr()

# Buat heatmap korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', vmin=-1, vmax=1, cbar=True, linewidths=0.5)
plt.title('Correlation between Math, Reading, and Writing Scores', fontsize=16)
plt.show()
````

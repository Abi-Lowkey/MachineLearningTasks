```python
# 2. Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Memuat Dataset
path = "/content/drive/MyDrive/Dataset/Position_Salaries.csv"
dataset = pd.read_csv(path)
```

```python
dataset.head()
```

```python
dataset
```

```python
# 4. Memisahkan fitur (Level) dan label (Salary)
X = dataset.iloc[:, 1:2].values  # Mengambil kolom Level
Y = dataset.iloc[:, 2].values    # Mengambil kolom Salary
```

```python
# 5. Fitting Linear Regression untuk perbandingan
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
# 6. Fitting Polynomial Regression
poly_reg = PolynomialFeatures(degree=4)  # Bisa ganti degree sesuai kebutuhan
X_poly = poly_reg.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
# 7. Visualisasi hasil Linear Regression
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# 8. Visualisasi hasil Polynomial Regression
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# 9. Prediksi dengan Linear Regression
level = 6.5  # Contoh level yang ingin diprediksi
print(f"Prediksi dengan Linear Regression untuk level {level}: {lin_reg.predict([[level]])[0]}")
```

```python
# 10. Prediksi dengan Polynomial Regression
print(f"Prediksi dengan Polynomial Regression untuk level {level}: {lin_reg_2.predict(poly_reg.fit_transform([[level]]))[0]}")
```
